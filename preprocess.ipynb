```python
import pandas as pd
```

```python
df = pd.read_csv("DATASET/rating.csv")
```

```python
print(df.head())
```

```python
# Since userIds are ordered from 1 to 7120, we need to convert them in a ZERO Based indexing format
df.userId -= 1
```

```python
print(df.head())
```

```python
#Store all unique movie ids
distinct_movie_ids = set(df.movieId.values)
```

```python
movie_index = {}
count = 0
for movie_id in distinct_movie_ids:
    movie_index[movie_id] = count
    count += 1
```

```python
print(movie_index)
```

```python
df = df.drop(columns=['timestamp'])
```

```python
print(df.head())
```

```python
df['movie_idx'] = df.apply(lambda row: movie_index[row.movieId],axis=1)
```

```python
print(df.head())
```

```python
df.to_csv("DATASET/edited_rating.csv")
```
