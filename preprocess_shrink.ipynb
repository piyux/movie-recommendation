```python
df = pd.read_csv("DATASET/edited_rating.csv")
```

```python
user_ids_count = Counter(df.userId)
movie_ids_count = Counter(df.movie_idx)
```

```python
user_ids_count
```

```python
movie_ids_count
```

```python
users_to_keep = int(1e4)
movies_to_keep = int(2e3)
```

```python
m = movies_to_keep
```

```python
user_ids = [ u for u, c in user_ids_count.most_common(users_to_keep)]
movie_ids = [m for m , c in movie_ids_count.most_common(m)]
```

```python
user_ids
```

```python
temp = set(movies_ids)
print(temp)
```

```python
df_small = df[df.userId.isin(user_ids) & df.movie_idx.isin(movie_ids)].copy()
```

```python
df_small
```

```python
new_user_id_dict = {}
i = 0
for old in user_ids:
    new_user_id_dict[old] = i
    i += 1
```

```python
new_user_id_dict
```

```python
new_movie_id_dict = {}
j = 0
for old in movie_ids:
    new_movie_id_dict[old] = j
    j += 1
```

```python
new_movie_id_dict
```

```python
#user ids and mmovie ids are now stored sequentially without any missing id in between
# setting of new ids -
df_small.loc[:,'userId'] = df_small.apply(lambda row: new_user_id_dict[row.userId],axis=1)
df_small.loc[:,'movie_idx'] = df_small.apply(lambda row: new_movie_id_dict[row.movie_idx],axis=1)
```
