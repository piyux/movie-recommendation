```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```python
df = pd.read_csv("DATASET/small_rating.csv")
```

```python
total_users , total_movies = df.userId.max() + 1 , df.movie_idx.max() + 1
```

```python
#split into train and test
df = shuffle(df)
cutoff = int(0.8 * len(df))
df_train = df.iloc[:cutoff]
df_test = df.iloc[cutoff:]
```

```python
#key-> user and value-> movies rated by him
user2movie = {}
#key -> movie and value-> users who rated it
movie2user = {}
#to look up ratings of 'j' movie rated by 'i' user
usermovie2rating ={}
```

```python
count = 0
```

```python
def update_user2movie_and_movie2user(row):
    global count
    count += 1
    if not count % 100000:
        print("processed: %.3f" % (float(count)/cutoff))
    i = int(row.userId)
    j = int(row.movie_idx)
    if i not in user2movie:
        user2movie[i] = [j]
    else:
        user2movie[i].append(j)
    if j not in movie2user:
        movie2user[j] = [i]
    else:
        movie2user[j].append(i)
    usermovie2rating[(i,j)] = row.rating
    
```

```python
df_train.apply(update_user2movie_and_movie2user,axis=1)
```

```python
usermovie2rating_test = {}
```

```python
count = 0
```

```python
def update_usermovie2rating_test(row):
    global count
    count += 1
    if not count % 100000:
        print("processed: %.3f" % (float(count)/len(df_test)))
    i = int(row.userId)
    j = int(row.movie_idx)
    usermovie2rating_test[(i,j)] = row.rating
    
```

```python
df_test.apply(update_usermovie2rating_test,axis=1)
```
